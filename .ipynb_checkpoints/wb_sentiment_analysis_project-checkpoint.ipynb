```python
from  nlp_toolbox import NLPTools
from wb_toolbox import WeiboSpiderTool
import tqdm
import pandas as pd
import os
```

```python
weibo = WeiboSpiderTool()
nlp = NLPTools()
```

```python
topic = '永州'
```

```python
# 观察爬取到的用户信息是否正常
user_weibo = pd.read_csv(r'E:\PycharmProjects\web_scraper\users\东方今报\1672519561.csv')
```

```python
# 利用我们事先定义好的事件关键词，对微博进行筛选
user_weibo_topic = user_weibo[user_weibo['正文'].str.contains(topic)]
```

```python
user_weibo_topic[['正文']]
```

```python
for bid in tqdm.tqdm(user_weibo_topic['bid'], desc='Weibo'):
    weibo.get_wb_comment(bid)
```

```python
comment_dir = r'E:\PycharmProjects\web_scraper\comment\yongzhou'
comments_ids = os.listdir(comment_dir)
```

```python
for wid in comments_ids:
    comments = pd.read_csv(os.path.join(comment_dir, comments_ids[wid]))
    sentiments = nlp.sentiment_predict(comments['评论内容'])
    comments['sentiments'] = sentiments
```
